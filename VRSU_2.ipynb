```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Load grayscale image
# -----------------------------
# NOTE: Change the path if needed
img = cv2.imread("/content/sample_data/download (2).jpg", 0)

if img is None:
    raise ValueError("Image not found! Check the file path.")

# -----------------------------
# Apply Otsu's Thresholding
# threshold() returns:
# 1) threshold value chosen by Otsu
# 2) binarized image
# -----------------------------
threshold_value, binary_img = cv2.threshold(
    img,
    0,                  # threshold value ignored when using OTSU
    255,                # max value
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)


print("Optimal Threshold Value:", threshold_value)

# -----------------------------
# Display results
# -----------------------------
plt.figure(figsize=(10, 4))

# Original Image
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis("off")

# Histogram
plt.subplot(1, 3, 2)
plt.title("Histogram")
plt.hist(img.ravel(), bins=256)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Otsu Binarized Image
plt.subplot(1, 3, 3)
plt.title("Otsu Binarized Image")
plt.imshow(binary_img, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
